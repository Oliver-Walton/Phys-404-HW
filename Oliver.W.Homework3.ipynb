```
# Calling libraries and defining forward and central difference schemes

import numpy as np
import matplotlib.pyplot as plt


def fwdiff(f , x , h=0.01): #forward diffenence scheme
    return (f(x+h)-f(x))/h

def cddiff(f , x , h=0.01): #central difference scheme
    return (f(x+h)-f(x-h))/(2*h)

cos = lambda x: np.cos(x)
sin = lambda x: np.sin(x)
```

```
print("Forward difference scheme at x=0: ", fwdiff(np.cos,0))
print("Central differnece scheme at x=0: ",cddiff(np.cos,0))
```

```
# Here we test the error on forward difference scheme as dx decreases. 

x0 = np.pi/4
theo = -np.sqrt(2)/2
dxarr = []
yarr = []

for i in range(1,17):
    dx = 10**(-i)
    y=fwdiff(np.cos,x0,dx)
    dxarr.append(dx)
    yarr.append(y)

plt.figure(figsize = [7,7])
plt.grid()
plt.xlabel("dx")
plt.ylabel("Error")
plt.plot(dxarr,abs(theo - yarr))
plt.xscale('log')
plt.yscale('log')
plt.show()
```

the best value for dx using forward difference scheme is dx = $10{-7}$ \
Giving an error of just under $10^{-7}$


```
#Here we test the error on central difference scheme as dx decreases.

x0 = np.pi/4
theo = -np.sqrt(2)/2
dxarr = []
yarr = []

for i in range(1,17):
    dx = 10**(-i)
    y=cddiff(np.cos,x0,dx)
    dxarr.append(dx)
    yarr.append(y)

plt.figure(figsize = [7,7])
plt.grid()
plt.xlabel("dx")
plt.ylabel("Error")   
plt.plot(dxarr,abs(theo - yarr))
plt.xscale('log')
plt.yscale('log')
plt.show()
```

the best value for dx using central difference scheme is dx = $10{-5}$ \
Giving an error of about $10^{-11}$
