```
###Part a) calculate all 4 solutions
import numpy as np

#variables
a = 1
b = 1
c = 10**-3

root = np.sqrt(b**2 - 4*a*c) #stuff under the sqrt

#firt quadratic equation
x1p = (-b + root)/2*a
x1n = (-b - root)/2*a

#second quadratic equation
x2p = (-2*c)/(b + root)
x2n = (-2*c)/(b - root)

print(x1p,x1n)
print(x2p,x2n)
```

```
### Part b) investigate errors 

#variables
a = 1
b = 1
c = 10**-1

#loop where c decreases by a factor of 10 every loop
for n in range(7):

    c = c*(10**-n)
    root = np.sqrt(b**2 - 4*a*c) #stuff under the sqrt

    #firt quadratic equation
    x1p = (-b + root)/2*a
    x1n = (-b - root)/2*a

    #second quadratic equation
    x2p = (-2*c)/(b + root)
    x2n = (-2*c)/(b - root)

    print(n+1)
    print(x1p)#pluses
    print(x2p)
    print(x1n)#negatives
    print(x2n)
```
