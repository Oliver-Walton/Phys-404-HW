```
### runga-kutta half year

import numpy as np
import matplotlib.pyplot as plt

G=1
M=1
tend=1*np.pi
h=0.01
t1=0
t2=0

#first mass
y1=[-0.5,0,0,-1/np.sqrt(2)]

#2nd mass
y2=[0.5,0,0,1/np.sqrt(2)]

def f(t, y):
    f=np.zeros(4)
    r=1
    f[0]=y[1]
    f[1]=-G*M/r/r/r*y[0]
    f[2]=y[3]
    f[3]=-G*M/r/r/r*y[2]
    return f
    
def rk4(t, h, y, n):
    k1 = np.zeros(n)
    k2 = np.zeros(n)
    k3 = np.zeros(n)
    k4 = np.zeros(n)
    k1=h*f(t,y)
    k2=h*f(t+0.5*h,y+0.5*k1)
    k3=h*f(t+0.5*h,y+0.5*k2) 
    k4=h*f(t+h,y+k3)  
    y+=(1/6)*(k1+2*k2+2*k3+k4) 
    return y

xp1=[]
yp1=[]
while(t1<tend):
    if((t1+h)>tend):
        h=tend-t1
    y1 = rk4(t1,h,y1,4) 
    xp1=np.append(xp1,y1[0])
    yp1=np.append(yp1,y1[2])
    t1 = t1+h

plt.plot(xp1,yp1)

xp2=[]
yp2=[]
while(t2<tend):
    if((t2+h)>tend):
        h=tend-t2
    y2 = rk4(t2,h,y2,4) 
    xp2=np.append(xp2,y2[0])
    yp2=np.append(yp2,y2[2])
    t2 = t2+h

plt.plot(xp2,yp2)
plt.show()
```
