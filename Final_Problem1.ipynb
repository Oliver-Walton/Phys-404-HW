1.Find the real roots of
\begin{equation}
x^4-6x^3+12.25x^2-11x+3.75=0
\end{equation}
using 1) bisection, and 2) Newton-Raphson methods.

```
#lets plot it first
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (x**4 - 6*x**3 + 12.25*x**2 - 11*x + 3.75)


x = np.arange(0,3.5,0.01)
y = f(x)
y2 = np.zeros(len(x))
plt.plot(x,y)
plt.plot(x,y2)
plt.show()

#roots appear to be at about 1 and 3. Notably there are 2 solutions
```

```
#Lets try bisection

import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (x**4 - 6*x**3 + 12.25*x**2 - 11*x + 3.75)

def bisection(xmin, xmax, f, Nmax = 100, eps = 1.0e-8): #this function was taken from Lect5-pre (slightly modified it)
    for i in range(0,Nmax):
        x=(xmin+xmax)/2.
        if(f(xmin)*f(x)>0):
            xmin = (xmin+xmax)/2 
        else:
            xmax = (xmin+xmax)/2 
        if(abs(f(x))<eps):
            print("Root is at",x," after ",i," iterations, with a precision of ",abs(f(x)))
            break
    return x

attempt1 = bisection(0,4,f)     #entire range
attempt2 = bisection(0,2.9,f)   #range excluding 3
attempt3 = bisection(1.1,4,f)   #range excluding 1

#appears to be 1 and 3
print("\nIf X=1, f(x) = ", f(1), "\nIf X=3, f(x) = ", f(3))
```

```
#Now lets try Newton-Raphson method

import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (x**4 - 6*x**3 + 12.25*x**2 - 11*x + 3.75)

def newtonraphson(x0, f,dx = 1.0e-6, Nmax = 100, eps = 1.0e-8): #this function was taken from Lect5-pre (slightly modified it)
    for i in range(0,Nmax):
        df = (f(x0+dx/2)-f(x0-dx/2))/dx # using central difference to calculate the derivative
        x0 = x0 - (f(x0)/df) 
        if(abs(f(x0))<eps):
            print("Root is at",x0," after ",i," iterations, with a precision of ",abs(f(x0)))
            break
    return x0

attempt1 = newtonraphson(0,f)
attempt2 = newtonraphson(1,f)
attempt3 = newtonraphson(2,f)
attempt4 = newtonraphson(3,f)
attempt5 = newtonraphson(4,f)

#appears to be 1 and 3 again
print("\nIf X=1, f(x) = ", f(1), "\nIf X=3, f(x) = ", f(3))

#roots are offically at 1 and 3 
```
