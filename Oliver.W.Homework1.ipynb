```
from numpy import *
import matplotlib.pyplot as plt
```

```
### All on the same graph

x=np.arange(-3.14,3.24,0.1)

# for n=11
fx=3.14/2
for n in np.arange(1,12,2):
    fx += 2*((np.sin(n*x))/n)
plt.plot(x,fx)

# for n=21
fx=3.14/2
for n in np.arange(1,22,2):
    fx += 2*((np.sin(n*x))/n)
plt.plot(x,fx)

# for n=31
fx=3.14/2
for n in np.arange(1,32,2):
    fx += 2*((np.sin(n*x))/n)
plt.plot(x,fx)

plt.show()
```
